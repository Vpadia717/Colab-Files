```python
import re

password = input('Enter Password : ')

pattern = r"[$&+,:;=?@#|'<>.-^*()%!,a-z,A-Z,0-9,.,.,.,.]"

if (re.match(pattern,password)) and (len(password) <= 8):
  print('Password Successfully Created')
else:
  print('Wrong password format')
```

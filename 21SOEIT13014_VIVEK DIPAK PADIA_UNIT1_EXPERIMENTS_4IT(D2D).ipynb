```python
def triangle(n):
  k = n - 1
  num = 65
  for i in range(0, n):
    for j in range(0, k):
      print(end=" ")
    k = k - 1
    for j in range(0, i+1):
        ch = chr(num)
        print(ch, end=" ")
    num = num + 1
    print("\r")

n = 5
triangle(n)
```

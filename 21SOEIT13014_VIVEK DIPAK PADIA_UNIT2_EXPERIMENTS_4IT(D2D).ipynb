```python
import mysql.connector
id = input("Enter Car id : ")
name = input("Enter Car name : ")
cname = input("Enter car company name : ")
ccost = input("Enter car cost : ")
tycar = input("Enter type of car : ")
caverage = input("Enter car average : ")
mydb = mysql.connector.connect(
    host='localhost', user='root', passwd='admin@123', database="RKU", port="3307")
print(mydb)

if (mydb):
    print("SuccessFully Connected")
else:
    print("Not Connected")

mycursor = mydb.cursor()
mycursor.execute("use RKU")
sql = "INSERT INTO Cars (sr_no, CarName, CarCompany, CarCost, TypeofCar, CarAverage) values(%s,%s,%s,%s,%s,%s)"
val = (id, name, cname, ccost, tycar, caverage)
try:
    mycursor.execute(sql, val)
    mydb.commit()
    print("data saved")
except:
    mydb.rollback()
mycursor.execute("SELECT * FROM Cars")
myresult = mycursor.fetchall()
for x in myresult:
    print(x)
```
